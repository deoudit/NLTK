```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("ratings.csv")
df.head()
```

```python
movie_titles = pd.read_csv("movies.csv")
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='movieId')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('white')
```

```python
%matplotlib inline
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating = pd.DataFrame(df.groupby('title')['rating'].mean())
rating.head()
```

```python
rating['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
rating.head()
```

```python
rating['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='num of ratings', data=rating, alpha=0.5)
```

```python
moviemat = df.pivot_table(index='userId', columns='title', values='rating')
```

```python
moviemat.head(10)
```

```python
rating.sort_values('num of ratings', ascending=False).head(10)
```

```python
Forrest_user_ratings = moviemat['Forrest Gump (1994)']
Redemption_user_ratings = moviemat['Shawshank Redemption, The (1994)']
```

```python
Forrest_user_ratings.head()
```

```python
similar_to_Forest = moviemat.corrwith(Forrest_user_ratings)
```

```python
similar_to_Redemption = moviemat.corrwith(Redemption_user_ratings)
```

```python
corr_Forest = pd.DataFrame(similar_to_Forest, columns=['Correlation'])
corr_Forest.dropna(inplace=True)
```

```python
corr_Forest.head()
```

```python
corr_Forest.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_Forest = corr_Forest.join(rating['num of ratings'])
```

```python
corr_Forest.head()
```

```python
corr_Forest[corr_Forest['num of ratings']>100].sort_values('Correlation',
                                                          ascending=False).head(10)
```

```python
corr_Redemption = pd.DataFrame(similar_to_Redemption, columns=['Correlation'])
corr_Redemption.dropna(inplace=True)
```

```python
corr_Redemption = corr_Redemption.join(rating['num of ratings'])
```

```python
corr_Redemption[corr_Redemption['num of ratings']>200].sort_values('Correlation',
                                                          ascending=False).head()
```
